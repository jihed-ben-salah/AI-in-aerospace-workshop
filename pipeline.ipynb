```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, roc_auc_score

# Load Dataset
data = pd.read_csv(".csv")

# Display first few rows
data.head()

# Data Exploration and Preprocessing
print(data.info())  # Check data types and missing values
```

```python
# Pairplot for mass, latitude, and longitude by meteorite type
sns.pairplot(data[['mass (g)', 'reclat', 'reclong', 'type']], hue='type', palette='coolwarm')
plt.suptitle('Pairplot of Meteorite Features by Type (Stony vs Iron)', y=1.02)
plt.show()

# Visualize the geographical distribution of meteorite landings (Stony vs Iron)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['reclong'], y=data['reclat'], hue=data['type'], palette='coolwarm')
plt.title('Geographical Distribution of Meteorite Landings')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
